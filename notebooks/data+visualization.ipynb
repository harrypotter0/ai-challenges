## Borrowers based on gender

```python
male_count = 0
female_count = 0
for i in kiva_loans['borrower_genders']:
    li = str(i).split(',')
    for j in li:
        if j.strip() == 'female':
            female_count += 1
        else:
            male_count += 1
```

```python
[male_count, female_count]
```

```python
sns.barplot(x=['male', 'female'], y=[male_count, female_count],
            label="Total", color="b")
```

Female borrowers are more than male borrowers.
